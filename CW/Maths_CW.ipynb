#  Question 1: Application of Differentiation in AI Model Optimization

```python
# variable assinging

x, y = symbols('x y')
```

```python
# calculkating loss fuction

loss = exp(x**2 + y**2) + cos(x*y)
loss
```

Question 1 - part (a)  - Calculating the Partial derivation


```python
# partial derivation

partial_x = diff(loss, x)
partial_y = diff(loss, y)


print("Partial derivative with respect to x:  ", partial_x)
print() # new line 
print("Partial derivative with respect to y:  ", partial_y)

```

 Question 1 part (b) - Calculatng the Gradient vector Fuction 


```python
# b. gradient Vector Fuction

from sympy import lambdify

gradient_function = lambdify((x,y), [partial_x,partial_y], 'numpy')

print("Gradient vector at (1,1):  " , gradient_function(1,1))
```

 Question 1 part(c) - Calculating the  3D surface plot


```python
# (c)  3d Surface plot

x_values = np.linspace(-2,2,100)
y_values = np.linspace(-2, 2 , 100)

X,Y = np.meshgrid(x_values, y_values)

loss_numpy = lambdify((x,y), loss, 'numpy')

Z = loss_numpy(X,Y)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
surface= ax.plot_surface(X,Y,Z, cmap="seismic")
ax.set_title('3D Surface Plot of Loss Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('loss(x,y)')
plt.show()
```

Question 1 Part (d) - Gradient Vector Significance 


```python
# Explanation for Gradient Vector Significance
discussion = """
Gradient vectors represent the direction and rate of the steepest ascent of a function.
In optimization, particularly in gradient descent methods, the negative gradient is used to iteratively update parameters in a machine learning model.

- The gradient vector points toward the maximum increase in the loss function.
- Gradient descent uses the opposite direction to minimize the function by taking steps proportional to the gradient.

For the given loss function:
L(x, y) = exp(x^2 + y^2) + cos(xy),

- The partial derivatives (L/x, L/y) indicate how the loss function changes with respect to x and y.
- The gradient vector guides us to reduce the loss iteratively during model training.
"""
print(discussion)
```

# **Question 3: Series Convergence and Numerical Approximation**

Question 3 part (a) -  Series Approximation of ln(1+x)

```python
# (a) Series Approximation of ln(1+x)

x_values = np.linspace(-1,1,100)
series_approximation = lambda x : sum ([(-1)**(n-1)* x **n/n for n in range (1,11)])
approximated_values = [series_approximation(x) for x in x_values]
actual_values = np.log(1+x_values)
actual_values
```

Question 3 part (b) - Ploting the apporximated and actual function


```python
# (b) Plotting the approximation and the actual function on the same graph
plt.plot(x_values, approximated_values, label="Approximation (10 terms)", color='r')
plt.plot(x_values, actual_values, label="Actual ln(1+x)", color='g', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Approximation of ln(1+x) using the first 10 terms of the series')
plt.legend()
plt.show()
```

Question 3 part (c) - Analysisng the convergence 


```python
errors = np.abs(np.array(approximated_values) - np.array(actual_values))
print("Maxium error is : ",np.max(errors))
print("The series converges well for |x|  1 but errors increase near the boundaries.")
plt.figure(figsize=(10, 6))
plt.plot(x_values, errors, label="Error (|Approximation - Actual|)", color="purple")
plt.title("Convergence Analysis of the Series Approximation", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("Error", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend(fontsize=12)
plt.grid()
plt.show()
```

#### Question 3 Part (c) - Convergence Analysis  - Disscussion part 

The series converges well nearx=0, where the error is minimal. 
As  approaches 1, the error increases due to slower convergence at the boundaries. 
The maximum error occurs near x=1 or x=1, as shown in the error plot.

# **Question 4: Fourier Transform for Data Filtering**

```python
from scipy.fft import fft, ifft, fftfreq
```

Question 4 part (a) - Noisy signal

```python
 # (a) nosiy signal

 t= np.linspace(0,1,500)
 s= np.sin(2*np.pi*t)+ 0.5 *np.sin(10*np.pi*t)+ 0.3 * np.sin(50*np.pi*t)

 # Fourier Transform

 freqs = fftfreq(len(t), d=(t[1]-t[0]))
 fft_result = fft(s)

plt.plot(freqs[:len(freqs)//2], np.abs(fft_result)[:len(fft_result)//2])
plt.title('Frequency Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

Question 4 Part (b) - Apply low-pass filter  

```python
fft_result[np.abs(freqs) > 20] = 0
filtered_signal = np.real(ifft(fft_result))

plt.plot(t, s, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal', linestyle='dashed')
plt.title('Low-Pass Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

### Question 4 (c) - Impact of Filtering  - Disscusion

The low-pass filter removes high-frequency noise above 20 Hz, leaving only the desired low-frequency components. In the frequency spectrum, the 25 Hz noise is removed, and the filtered signal becomes smoother. This technique is crucial in real-world applications like audio processing, where it helps reduce noise and improve signal clarity.

# **Question 5: Image Processing Using Fourier and DCT**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, ifft2, fftshift
from scipy.fftpack import dct, idct
from scipy.ndimage import gaussian_filter
from numpy.fft import ifftshift



# Loading image
image = cv2.imread('Image.png',0)


```

Question 5 Part  (a) - Edge detection using 2D Fourier Transform

```python


f_transform = np.fft.fft2(image)
f_shift=np.fft.fftshift(f_transform)
magitude_spectrum =20*np.log(np.abs(f_shift))

# Applying a high pass filterr to enchances edges
rows,cols=image.shape
crow,ccol=rows//2, cols//2
mask=np.ones((rows,cols),np.uint8)

# Radius for the high pass filter

radius = 30

#setting a circle region to 0
cv2.circle(mask,(ccol,crow), radius,0,-1)

# Applying the hihg pass filter in the Forier domain

filtered_f_shift = f_shift * mask

# Performing the inverse Fourier Transform to visualize edges
inverse_f_shift = np.fft.ifftshift(filtered_f_shift)
filtered_image = np.abs(np.fft.ifft2(inverse_f_shift))


# Displaying the detected edges
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title("Edges Detected with Fourier Transform")
plt.show()
```

Question 5 part (b) -  Gaussian blur using 2D Fourier Transform

```python


# Dimensions of the image
rows, cols = image.shape

# Sigma value for Gaussian blur
sigma = 30

### 1. Applying Gaussian Blur in the Frequency Domain
# Perform Fourier Transform and shift zero-frequency component to the center
f_transform = fft2(image)
f_shift = fftshift(f_transform)

# Create a Gaussian mask in the frequency domain
x = np.linspace(-cols // 2, cols // 2, cols)
y = np.linspace(-rows // 2, rows // 2, rows)
X, Y = np.meshgrid(x, y)
gaussian_mask = np.exp(-(X**2 + Y**2) / (2 * sigma**2))

# Apply the Gaussian mask in the frequency domain
blurred_f_shift = f_shift * gaussian_mask

# Perform the inverse Fourier Transform to get the blurred image
inverse_blurred_f_shift = ifftshift(blurred_f_shift)  # Shift back the zero-frequency component
blurred_image_fourier = np.abs(ifft2(inverse_blurred_f_shift))  # Reconstruct the image

### 2. Applying Gaussian Blur in the Spatial Domain

kernel_size = (21, 21)
blurred_image_standard = cv2.GaussianBlur(image, kernel_size, sigmaX=sigma)


plt.figure(figsize=(12, 6))

# Fourier-domain Gaussian blur
plt.subplot(1, 2, 1)
plt.imshow(blurred_image_fourier, cmap='gray')
plt.title("Blurred with Fourier Transform")
plt.axis("off")

# Spatial-domain Gaussian blur
plt.subplot(1, 2, 2)
plt.imshow(blurred_image_standard, cmap='gray')
plt.title("Blurred with Standard Gaussian Blur")
plt.axis("off")

plt.tight_layout()
plt.show()
```

Question 5 part (c) Scaling with DCT

```python


# Apply 2D DCT
def dct2(image):
    return dct(dct(image.T, norm='ortho').T, norm='ortho')

# Apply 2D inverse DCT
def idct2(image):
    return idct(idct(image.T, norm='ortho').T, norm='ortho')

# Apply the 2D Discrete Cosine Transform (DCT)
image_dct = dct2(image)

# Scaling down the image to 240px X 240px using DCT coefficients
scaled_size = 240
scaled_dct = image_dct[:scaled_size, :scaled_size]

# Zero-pad the scaled DCT coefficients back to the original size
padded_dct = np.zeros_like(image_dct)
padded_dct[:scaled_size, :scaled_size] = scaled_dct

# Perform the inverse DCT to reconstruct the scaled image
scaled_image = idct2(padded_dct)

# Normalize the scaled image to the range [0, 255] for visualization
scaled_image_normalized = np.clip(scaled_image, 0, 255).astype(np.uint8)

# Show original and scaled images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Scaled Image (240x240) Using DCT")
plt.imshow(scaled_image_normalized, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()



```

Question 5 part (d)-  Compression artifacts

```python
# Applying the aggressive DCT compression by keeping only a small number of coefficients
compression_level = 50
compressed_dct = np.zeros_like(image_dct)
compressed_dct[:compression_level, :compression_level] = image_dct[:compression_level, :compression_level]

# Reconstructing the image from the compressed DCT
image_compressed = idct(idct(compressed_dct.T, norm='ortho').T, norm='ortho')
image_compressed_normalized = np.clip(image_compressed, 0, 255).astype(np.uint8)

# Visualizing the artifacts (ringing and blocking)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Compressed Image (DCT Artifacts)")
plt.imshow(image_compressed_normalized, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Difference from Original (Highlighting Artifacts)")
difference = np.abs(image - image_compressed_normalized)
plt.imshow(difference, cmap='hot')
plt.axis("off")

plt.show()
```

#### Question 5 part (d) - discussion part 

Aggressive DCT compression introduces two main artifacts: blocking and ringing. Blocking appears as visible grid-like patterns due to the loss of continuity between independently processed blocks. Ringing manifests as ripples or oscillations around edges, caused by discarding high-frequency details. The difference heatmap highlights these distortions, with the most significant impact around sharp edges and fine details.
